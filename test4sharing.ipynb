```python
from IPython.display import HTML

HTML('''<script>
    var code_show=false; //true -> hide code at first

    function code_toggle() {
        $('div.prompt').hide(); // always hide prompt

        if (code_show){
            $('div.input').hide();
        } else {
            $('div.input').show();
        }
        code_show = !code_show
    }
    $( document ).ready(code_toggle);
</script>''')
#<a href="javascript:code_toggle()">[Toggle Code]</a>''')
```
